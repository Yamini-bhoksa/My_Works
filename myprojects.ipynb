# 1. Basic Data Analysis (Titanic Dataset)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset('titanic')
print(df.head())

# Basic info
print(df.info())

# Survival rate
print(df['survived'].value_counts(normalize=True))

# Visualization
sns.countplot(x='sex', hue='survived', data=df)
plt.title('Survival Rate by Gender')
plt.show()
```

# 2. Student Marks Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Name': ['A', 'B', 'C', 'D', 'E'],
        'Maths': [85, 70, 90, 65, 88],
        'Science': [80, 75, 85, 70, 92],
        'English': [78, 72, 89, 68, 85]}

df = pd.DataFrame(data)
df['Average'] = df[['Maths', 'Science', 'English']].mean(axis=1)
print(df)

plt.bar(df['Name'], df['Average'])
plt.title('Average Marks of Students')
plt.show()
```

# 3. Weather Data Analysis

```python
import pandas as pd

data = {'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
        'Temp': [32, 35, 30, 28, 33],
        'Rainfall': [2.0, 0.0, 1.2, 3.5, 0.0]}
df = pd.DataFrame(data)

print("Average Temperature:", df['Temp'].mean())
print("Days with Rain:", (df['Rainfall'] > 0).sum())
```

# 4. Covid-19 Data Tracker

```python
import pandas as pd
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/datasets/covid-19/main/data/countries-aggregated.csv'
df = pd.read_csv(url)

india = df[df['Country'] == 'India']
plt.plot(india['Date'], india['Confirmed'], label='Confirmed Cases')
plt.plot(india['Date'], india['Deaths'], label='Deaths')
plt.legend()
plt.title('COVID-19 India Trend')
plt.xlabel('Date')
plt.ylabel('Count')
plt.show()
```

# 5. Movie Ratings Analysis

```python
import pandas as pd
import seaborn as sns

data = {'Movie': ['A', 'B', 'C', 'D', 'E'],
        'Genre': ['Action', 'Drama', 'Action', 'Comedy', 'Drama'],
        'Rating': [7.8, 8.2, 6.9, 7.5, 8.0]}
df = pd.DataFrame(data)

avg = df.groupby('Genre')['Rating'].mean()
print(avg)

sns.barplot(x=avg.index, y=avg.values)
```

# 6. Sales Data Analysis

```python
import pandas as pd

sales = {'Month': ['Jan', 'Feb', 'Mar', 'Apr'],
         'Sales': [2000, 2500, 2200, 2700]}
df = pd.DataFrame(sales)

print("Total Sales:", df['Sales'].sum())
print("Best Month:", df.loc[df['Sales'].idxmax(), 'Month'])
```

## 7. Stock Price Visualization

```python
import yfinance as yf
import matplotlib.pyplot as plt

stock = yf.download('AAPL', start='2024-01-01', end='2024-06-30')
plt.plot(stock['Close'])
plt.title('Apple Stock Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

# 8. Word Frequency Counter

```python
from collections import Counter

text = "Python is great and Python is fun. I love Python!"
words = text.lower().split()
count = Counter(words)

print(count)
```
